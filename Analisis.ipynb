```python
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn
import matplotlib as plt 


%matplotlib inline
```

```python
ecoData = pd.read_csv("ecoData.csv", index_col = 0)
ecoData.head()
```

```python
ecoData = ecoData.sample(100000)
ecoData = ecoData.dropna()
ecoData = ecoData.reset_index()
```

```python

a4_dims = (11.7, 8.27)
fig, ax = plt.pyplot.subplots(figsize=a4_dims)

for i in ecoData["day_Week"].unique():
    sns.distplot( ecoData["travel_label_5min"][ecoData["day_Week"] == i], label=i)

sns_plot =sns.distplot( ecoData["travel_label_5min"] )
ax.set(xlabel='SEGUNDOS', ylabel='FREQUENCIA')
ax.set_title("DISTRIBUCION POR DIA DE LA SEMANA")

sns_plot.get_figure().savefig("Distribucion2.png")
```
